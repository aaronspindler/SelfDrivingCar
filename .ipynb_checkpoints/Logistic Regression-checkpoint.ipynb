```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_points = 100
np.random.seed(0)
top_region = np.array([np.random.normal(10, 2, num_points), np.random.normal(12, 2, num_points)]).transpose()
bottom_region = np.array([np.random.normal(5, 2, num_points), np.random.normal(6, 2, num_points)]).transpose()
_, ax = plt.subplots(figsize=(4,4))
ax.scatter(top_region[:, 0], top_region[:, 1], color='r')
ax.scatter(bottom_region[:, 0], bottom_region[:, 1], color='b')
plt.show()
```
